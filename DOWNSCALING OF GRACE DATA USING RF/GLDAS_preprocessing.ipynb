Inverse Distance Weighting (IDW)  #krigging

```python
import numpy as np
import pandas as pd
from scipy.spatial import cKDTree

# Sample DataFrame (Ensure it has 'coordinates' as tuples)
df_merged[['longitude', 'latitude']] = pd.DataFrame(df_merged['coordinates'].tolist(), index=df_merged.index)

# Separate known and unknown data points
known = df_merged[df_merged['CanopInt_inst_mean'].notna()]
unknown = df_merged[df_merged['CanopInt_inst_mean'].isna()]

# Build KDTree for spatial search
tree = cKDTree(known[['longitude', 'latitude']].values)

# Define IDW function
def idw_interpolation(x, y, known_points, known_values, k=5, power=2):
    """Inverse Distance Weighting (IDW) interpolation"""
    distances, idx = tree.query([[x, y]], k=k)
    distances = distances.flatten()
    idx = idx.flatten()
    
    # Avoid division by zero
    distances = np.where(distances == 0, 1e-10, distances)
    
    weights = 1 / distances ** power
    interpolated_value = np.sum(weights * known_values[idx]) / np.sum(weights)
    
    return interpolated_value

# Apply IDW to fill NaN values
df_merged.loc[df_merged['CanopInt_inst_mean'].isna(), 'CanopInt_inst_mean'] = \
    unknown.apply(lambda row: idw_interpolation(row['longitude'], row['latitude'], 
                                                known[['longitude', 'latitude']].values, 
                                                known['CanopInt_inst_mean'].values), axis=1)

df_merged
```
